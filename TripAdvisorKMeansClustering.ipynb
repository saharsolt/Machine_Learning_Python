```python
import pandas as pd
data = pd.read_csv('tripadvisor-data.csv')
data.head(10)
```

```python
import matplotlib.pyplot as plt

data = data.values
plt.figure()
plt.scatter(data[:,0],data[:,1], marker='.', color='blue')
plt.xlabel("Art galleries")
plt.ylabel("Religious institutions")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(init='random', n_clusters=2, n_init=10)
kmeans_model.fit(data)
```

```python
print("The lowest SSE= {:.4f}".format(kmeans_model.inertia_))
print("The location of the centroid is {}".format(kmeans_model.cluster_centers_))
print("The number of iterations required to converge=",kmeans_model.n_iter_)
```

```python
print(kmeans_model.labels_[:10])
```

```python
u = kmeans_model.cluster_centers_

labels = kmeans_model.labels_
plt.figure()
plt.scatter(data[labels==0,0],data[labels==0,1],marker='.',color='blue', label='First cluster')
plt.scatter(data[labels==1,0],data[labels==1,1],marker='.',color='orange', label='Second cluster')
plt.scatter(u[:,0],u[:,1],marker='*',color='black', label='Cluster center')
plt.xlabel('Art galleries')
plt.ylabel('Religious institutions')
plt.legend()
plt.show()
```

```python
kmeans_newmodel= KMeans(init='random', n_clusters=3, n_init=10)
kmeans_newmodel.fit(data)
```

```python
u_new = kmeans_newmodel.cluster_centers_
labels_new= kmeans_newmodel.labels_
plt.figure()
plt.scatter(data[labels_new==0,0],data[labels_new==0,1],marker='.',color='blue', label='First cluster')
plt.scatter(data[labels_new==1,0],data[labels_new==1,1],marker='.',color='orange', label='Second cluster')
plt.scatter(data[labels_new==2,0],data[labels_new==2,1],marker='.',color='green', label='Third cluster')
plt.scatter(u_new[:,0],u_new[:,1],marker='*',color='black', label='Cluster center')
plt.xlabel('Art galleries')
plt.ylabel('Religious institutions')
plt.legend()
plt.show()
```
