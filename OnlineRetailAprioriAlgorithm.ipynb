```python
df = df.dropna(subset=['Description'],axis=0)
```

```python
df.isnull().sum()
```

```python
df = df[~df['InvoiceNo'].apply(lambda x: str(x)[0].isalpha())]
```

```python
df['Description']= df['Description'].apply(lambda x: str(x).strip())
```

```python
display(df['Description'])
```

```python
baskets = df.groupby(['InvoiceNo'])
list_baskets = baskets['Description'].apply(list)
print(list_baskets)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

trans_encod = TransactionEncoder()
trans_arr = trans_encod.fit(list_baskets).transform(list_baskets)
df_encod = pd.DataFrame(trans_arr, columns = trans_encod.columns_, index= list_baskets.index)
df_encod
```

```python
freq_item = apriori(df_encod, min_support=0.025,use_colnames=True)
freq_item
```

```python
association_rules(freq_item, metric='confidence', min_threshold=0.7)
```
