```python
import pandas as pd
data = pd.read_csv('clustering-data.csv')
data.head()
```

```python
from sklearn.decomposition import PCA
pca_model = PCA(n_components=2)
X_pca = pca_model.fit_transform(data)
print("Shape of data: {}".format(data.shape))
print("Shape of data after PCA: {}".format(X_pca.shape))
```

```python
print(pca_model.explained_variance_)
print(pca_model.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X_pca[:,0],X_pca[:,1], marker='.',color='orange')
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.title('Two features after PCA')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
kmean3_model = KMeans(init='random', n_clusters=3, n_init=10)
kmean3_model.fit(X_pca)
```

```python
u = kmean3_model.cluster_centers_
label = kmean3_model.labels_  
plt.figure()
plt.scatter(X_pca[label==0,0],X_pca[label==0,1],color='blue',marker='.',label='First cluster')
plt.scatter(X_pca[label==1,0],X_pca[label==1,1],color = 'orange',marker='.',label='Second cluster')
plt.scatter(X_pca[label==2,0],X_pca[label==2,1],color='green',marker='.',label='Third cluster')
plt.scatter(u[:,0],u[:,1],marker='*',color='black', label='Cluster centroid')
plt.title("Clustering data with k=3")
plt.legend()
plt.show()
```

```python
print("The number of iterations required to converge", kmean3_model.n_iter_)
print("The location of the centroid \n", kmean3_model.cluster_centers_)
print("The lowest SSE", kmean3_model.inertia_)
print("Labels",kmean3_model.labels_)
```

```python
from sklearn.cluster import KMeans
kmean4_model = KMeans(init='random', n_clusters=4, n_init=4)
kmean4_model.fit(X_pca)
```

```python
u4 = kmean4_model.cluster_centers_
label4 = kmean4_model.labels_
print(u4)
print(label4)
```

```python
plt.figure()
plt.scatter(X_pca[label4==0,0],X_pca[label4==0,1],color='blue',marker='.',label='First cluster')
plt.scatter(X_pca[label4==1,0],X_pca[label4==1,1],color = 'green',marker='.',label='Second cluster')
plt.scatter(X_pca[label4==2,0],X_pca[label4==2,1],color='red',marker='.',label='Third cluster')
plt.scatter(X_pca[label4==3,0],X_pca[label4==3,1],color='orange',marker='.',label='Fourth cluster')
plt.scatter(u4[:,0],u4[:,1],marker='*',color='black', label='Cluster centroid')
plt.title("Clustering data with k=4")
plt.legend()
plt.show()
```
