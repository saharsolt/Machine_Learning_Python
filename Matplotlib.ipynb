```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.plot(x,y_sin,label="Sin")
plt.plot(x,y_cos,label="Cos")
plt.legend()

plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title("Sin and Cos Functions")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x)
y_linear=0.5*x+1

plt.plot(x,y_sin, label='Sin', color='Red', linewidth=3.5)
plt.plot(x,y_linear, label='Linear', linestyle='--')
plt.legend()

plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Sin and Linear Functions')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.figure(figsize=(6,7))
plt.subplot(2,1,2)

plt.plot(x,y_sin, label='Sin', color='blue')
plt.title('Making subplot')

plt.subplot(2,1,1)
plt.plot(x,y_cos, label='Cos', color='black' , linestyle='--')

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data1=np.random.random((50,2))
data2=np.random.random((50,2))

fig,ax = plt.subplots(figsize=(6,6))
ax.scatter(data1[:,0],data1[:,1], color='orange',linewidth=4,marker='.')
ax.scatter(data2[:,0],data2[:,1], color='blue',linewidth=4,marker='*')


ax.set_xlabel('x',fontsize=12)
ax.set_ylabel('y',fontsize=12)
ax.set_title('An example of scatter plot',fontsize=14)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

fig,ax = plt.subplots(2,figsize=(6,7))

ax[0].plot(x,y_sin, label='Sin', color='blue')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

ax[1].plot(x,y_cos, label='Cos', color='black' , linestyle='--')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].legend()

ax[0].set_title('Making OOI for Sin')
ax[1].set_title('Making OOI for Cos')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10,0.5)

fig,ax = plt.subplots(2,2)

for i in range(2):
    for j in range(2):
        y=2*(i+j)*x+1
        ax[i,j].plot(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,10,100)
y = -0.5*x**2 + 2*x - 3

plt.plot(x,y,label='curve',color='red', linewidth=2, linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y versus x')
plt.legend()

data_rnd_1=np.random.random((100,2))
data_rnd_2=np.random.random((120,2))

fig,ax = plt.subplots(2,figsize=(6,9))
ax[0].scatter(data_rnd_1[:,0],data_rnd_1[:,1],marker='*', color='green')
ax[1].scatter(data_rnd_2[:,0],data_rnd_2[:,1],marker='.', color='red')

ax[0].set_xlabel('dimension 1')
ax[0].set_ylabel('dimension 2')
ax[1].set_xlabel('dimension 1')
ax[1].set_ylabel('dimension 2')

plt.show()
```
